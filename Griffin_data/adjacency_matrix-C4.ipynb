# User dataframe

```python
nodes_C1 = pd.read_csv('nodes_C4_1.csv')
nodes_C2 = pd.read_csv('nodes_C4_2.csv')
nodes_C0 = pd.read_csv('nodes_C4_0.csv')
nodes_C3 = pd.read_csv('nodes_C4_3.csv')

node_files = [nodes_C2, nodes_C0, nodes_C3]

time_series = nodes_C1.append(node_files, ignore_index=True)
time_series
```

```python
# adjusted
clean_df = df[['src', 'dst', 'start_time_', 'messages_sent', 'messages_received', 'messages_total', 
               'contribution_index', 'sentiment_avg', 'emotionality_avg', 'complexity_avg']]
clean_df.rename(columns={'start_time_':'date', 'messages_sent':'mes_sent', 'messages_received':'mes_received',
                         'messages_total':'mes_total', 'contribution_index':'contr_index', 
                         'emotionality_avg':'emoti_avg', 'complexity_avg':'compl_avg'}, inplace=True)
clean_df['neighbours'] = 0
clean_df['neighbours_cluster'] = 0
clean_df['neighbours_!cluster'] = 0

clean_df.sort_values(['src', 'date'], inplace=True)
clean_df = clean_df.merge(time_series[['id', 'cluster', 'twitter_query']], left_on='dst', right_on='id', how='left').drop('id', axis=1).rename(columns={'cluster':'cluster_dst', 'twitter_query':'twitter_query_dst'})
clean_df = clean_df.merge(time_series[['id', 'cluster', 'twitter_query']], left_on='src', right_on='id', how='left').drop('id', axis=1).rename(columns={'cluster':'cluster_src', 'twitter_query':'twitter_query_src'})

clean_df['cluster_equal'] = clean_df['cluster_src'].eq(clean_df['cluster_dst'])
clean_df['query_equal'] = clean_df['twitter_query_src'].eq(clean_df['twitter_query_dst'])
clean_df
```

```python
clean_df.cluster_src.unique()
```

```python
for index, row in tqdm(clean_df.iterrows(), total=clean_df.shape[0]):
    user = clean_df['src'][index]
    clean_df.loc[(clean_df['src'] == user, 'neighbours')] = len(clean_df[clean_df['src'] == user])
    clean_df.loc[(clean_df['cluster_equal'] == True) & (clean_df['src'] == user), 'neighbours_cluster'] = len(clean_df.loc[((clean_df['cluster_equal'] == True) & (clean_df['src'] == user))])
    clean_df.loc[(clean_df['cluster_equal'] == False) & (clean_df['src'] == user), 'neighbours_!cluster'] = len(clean_df.loc[((clean_df['cluster_equal'] == False) & (clean_df['src'] == user))])
```

```python
clean_df.to_csv('Adjacency_Matrix_Output_C4.csv')
```

# Preprocessing

```python
adj_mat_edit = adj_mat.copy()
```

```python
adj_mat_edit['neighbours_!cluster_edit'] = 0
adj_mat_edit['neighbours_cluster_edit'] = 0
```

```python
adj_mat_edit.loc[(adj_mat_edit['neighbours_cluster'] != 0), 'neighbours_!cluster_edit'] = adj_mat_edit.loc[(adj_mat_edit['neighbours_cluster'] != 0), 'neighbours'] - adj_mat_edit.loc[(adj_mat_edit['neighbours_cluster'] != 0), 'neighbours_cluster'] 
adj_mat_edit.loc[(adj_mat_edit['neighbours_cluster'] == 0), 'neighbours_!cluster_edit'] = adj_mat_edit.loc[(adj_mat_edit['neighbours_cluster'] == 0), 'neighbours_!cluster']
```

```python
col_num = ["neighbours", "neighbours_cluster", "neighbours_!cluster", "neighbours_cluster_edit", "neighbours_!cluster_edit"]
adj_mat_edit[col_num] = adj_mat_edit[col_num].apply(pd.to_numeric)
```

```python
adj_mat_edit.loc[(adj_mat_edit['neighbours_!cluster'] == 0), 'neighbours_cluster_edit'] = adj_mat_edit.loc[(adj_mat_edit['neighbours_!cluster'] == 0), 'neighbours'] - adj_mat_edit.loc[(adj_mat_edit['neighbours_!cluster'] == 0), 'neighbours_!cluster_edit']
adj_mat_edit.loc[(adj_mat_edit['neighbours_cluster'] == 0), 'neighbours_cluster_edit'] = adj_mat_edit.loc[(adj_mat_edit['neighbours_cluster'] == 0), 'neighbours'] - adj_mat_edit.loc[(adj_mat_edit['neighbours_cluster'] == 0), 'neighbours_!cluster_edit']
```

```python
adj_mat_edit = adj_mat_edit.drop(columns=['neighbours_cluster', 'neighbours_!cluster'])
adj_mat_edit = adj_mat_edit.rename(columns={'neighbours_!cluster_edit':'neighbours_!cluster',
                                            'neighbours_cluster_edit':'neighbours_cluster'})
```

```python
adj_mat_edit['date'] = pd.to_datetime(adj_mat_edit['date'])
adj_mat_edit = adj_mat_edit.sort_values('date').set_index('date')
adj_mat_edit = adj_mat_edit.loc['2021-10-24':'2021-12-26']
adj_mat_edit
```
